```python
## get AUCs
def get_auc(feature, nbins = 10):
    #print(feature)
    range_val = (np.nanpercentile(df[feature],5e-3),
                                          np.nanpercentile(df[feature],100-5e-3))
    #print(range_val)
    ctrlsample = df[df.gene == 'nontargeting'].sample(n=50000,random_state=0)
    
   
    ctrl = np.cumsum(np.histogram(ctrlsample[feature],bins=nbins,range=range_val)[0]
                   /ctrlsample[feature].shape[0])

    aucs = (df.groupby(['gene','cell_barcode_0'])[feature].apply(
           lambda x: np.trapz((np.cumsum(np.histogram(x, bins=nbins,
            range=range_val)[0]/len(x)) - ctrl), dx = 1) / nbins))
    
    return aucs
```
